```python
import pandas as pd
import random
from NetEmbs import *
def dirty_data(N=2, tax_rates=[0.1, 0.25], types=['OK', 'NaN', 'Zeros', 'VeryLow', 'String', 'Negative', 'VeryBig'], 
               titles=["ID", "FA_Name", "Journal", "Date", "Value"]):
    cur_id = 0
    data = pd.DataFrame(columns=titles)
    for tax in tax_rates:
        for _ in range(N):
            for dirty_type in types:
                # generate amounts
                if dirty_type == "OK":
                    rev = -random.randint(10, 1000)
                    t = rev * tax
                    tr = -(rev + t)
                if dirty_type == "NaN":
                    rev = None
                    t = float('nan')
                    tr = None
                if dirty_type == "Zeros":
                    rev = 0
                    t = 0*tax
                    tr = 0
                if dirty_type == "VeryLow":
                    rev = -2**-1025
                    t = rev*tax
                    tr = -(rev+t)
                if dirty_type == "VeryBig":
                    rev = -2**1022
                    t = rev*tax
                    tr = -(rev+t)
                if dirty_type == "String":
                    rev = "dirty_value"
                    t = -10.0
                    tr = 11.0
                if dirty_type == "Negative":
                    rev = random.randint(10, 1000)
                    t = -rev * tax
                    tr = rev + t
                data = data.append(
                    [pd.Series([cur_id, "Revenue", "Sales ledger", "01/01/2017", rev], index=data.columns),
                     pd.Series([cur_id, "Tax", "Sales ledger", "01/01/2017", t], index=data.columns),
                     pd.Series([cur_id, "Trade Receivables", "Sales ledger", "01/01/2017", tr],
                               index=data.columns)]
                    , ignore_index=True)
                cur_id += 1
    return data
def dirty_dataM(N=2, tax_rates=[0.1, 0.25], types=['OK', 'NaN', 'Zeros', 'VeryLow', 'String', 'Negative', 'VeryBig'], 
               titles=["ID", "FA_Name", "Journal", "Date", "amount", "type"]):
    cur_id = 0
    data = pd.DataFrame(columns=titles)
    for tax in tax_rates:
        for _ in range(N):
            for dirty_type in types:
                # generate amounts
                if dirty_type == "OK":
                    rev = random.randint(10, 1000)
                    t = rev * tax
                    tr = rev + t
                if dirty_type == "NaN":
                    rev = None
                    t = float('nan')
                    tr = None
                if dirty_type == "Zeros":
                    rev = 0
                    t = 0*tax
                    tr = 0
                if dirty_type == "VeryLow":
                    rev = 2**-1025
                    t = rev*tax
                    tr = rev+t
                if dirty_type == "VeryBig":
                    rev = 2**1022
                    t = rev*tax
                    tr = rev+t
                if dirty_type == "String":
                    rev = "dirty_value"
                    t = -10.0
                    tr = 11.0
                if dirty_type == "Negative":
                    rev = random.randint(10, 1000)
                    t = -rev * tax
                    tr = rev + t
                data = data.append(
                    [pd.Series([cur_id, "Revenue", "Sales ledger", "01/01/2017", rev, "credit"], index=data.columns),
                     pd.Series([cur_id, "Tax", "Sales ledger", "01/01/2017", t, "credit"], index=data.columns),
                     pd.Series([cur_id, "Trade Receivables", "Sales ledger", "01/01/2017", tr, "debit"],
                               index=data.columns)]
                    , ignore_index=True)
                cur_id += 1
    return data
```

```python
def countStrings(df, col=["Value"]):
    output = dict()
    for title in col:
        output[title] = df[title].map(lambda x: 1 if type(x)==str else 0).sum()
    return output
def countNaN(df, col=["Value"]):
    output = dict()
    for title in col:
        output[title] = df[title].isnull().sum()
    return output
def countDirtyData(df, col=["Value"]):
    print("Strings in numeric columns: ", countStrings(df, col))
    print("NaN in numeric columns: ", countNaN(df, col))    
```

```python
import pandas as pd
import random
def zerosData(N=2, taxes=[0.06, 0.25], titles=["ID", "FA_Name", "Journal", "Date", "Debit", "Credit"]):
    if taxes is not None:
        tax_rates = taxes
    if N is not None:
        N = 10
    cur_id = 0
    data = pd.DataFrame(columns=titles)
    for tax in tax_rates:
        for _ in range(N):
            # generate amounts
            rev = random.randint(10, 1000)
            t = rev * tax
            tr = rev + t
            data = data.append(
                [pd.Series([cur_id, "Revenue", "Sales ledger", "01/01/2017", 0.0, rev], index=data.columns),
                 pd.Series([cur_id, "Tax", "Sales ledger", "01/01/2017", 0.0, t], index=data.columns),
                 pd.Series([cur_id, "Trade Receivables", "Sales ledger", "01/01/2017", tr, 0.0],
                           index=data.columns)]
                , ignore_index=True)
            cur_id += 1
            data = data.append([pd.Series([cur_id, "Trade Receivables", "Sales ledger", "01/01/2017", 0.0, tr],
                                          index=data.columns),
                                pd.Series([cur_id, "Cash", "Sales ledger", "01/01/2017", tr, 0.0],
                                          index=data.columns)]
                               , ignore_index=True)
            cur_id += 1
            data = data.append(
                [pd.Series([cur_id, "Revenue", "Sales ledger", "01/01/2017", 0.0, 0.0], index=data.columns),
                 pd.Series([cur_id, "Tax", "Sales ledger", "01/01/2017", 0.0, 0.0], index=data.columns),
                 pd.Series([cur_id, "Trade Receivables", "Sales ledger", "01/01/2017", 0.0, 0.0],
                           index=data.columns)]
                , ignore_index=True)
            cur_id += 1
            data = data.append([pd.Series([cur_id, "Trade Receivables", "Sales ledger", "01/01/2017", 0.0, 0.0],
                                          index=data.columns),
                                pd.Series([cur_id, "Cash", "Sales ledger", "01/01/2017", 0.0, 0.0],
                                          index=data.columns)]
                               , ignore_index=True)
            cur_id += 1

    return data
```
